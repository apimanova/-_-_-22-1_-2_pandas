```python
import numpy as np
import pandas as pd
```

```python
recipes = pd.read_csv('recipes_sample.csv')
reviews = pd.read_csv('reviews_sample.csv')
reviews.rename(columns={'Unnamed: 0': 'indexes'}, inplace=True)
recipes['submitted']= pd.to_datetime(recipes['submitted'])
```

```python
print('Пропуски по столбцам в recipes\n', recipes.isnull().sum(), '\n')
print('Пропуски по столбцам rewiews \n', reviews.isnull().sum(), '\n')
a = len(recipes.index)-np.count_nonzero(recipes.isnull().sum(axis=1)) # Строки с пропусками recipes
b = len(reviews.index)-np.count_nonzero(reviews.isnull().sum(axis=1)) # Строки с пропусками reviews
print('Доля строк, содержащих пропуски в recipes:', a / len(recipes.index))
print('Доля строк, содержащих пропуски в reviews:', b / len(reviews.index))
```

### Работа со строковыми данными в `pandas`

3.1  Добавьте в таблицу `recipes` столбец `description_length`, в котором хранится длина описания рецепта из столбца `description`.

3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
noWS = recipes['name'].str.replace(' ', '')
recipes['name_word_count'] = noWS.str.findall(r'[A-Z]').str.len()
recipes.to_csv('3_3.csv')
display(recipes)
```

### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
idGroup = recipes.groupby(recipes['contributor_id'].tolist(), as_index= False).size()
idGroup.rename(columns = {'index':'contributor_id'}, inplace = True)
display('Количество рецептов, представленных каждым из участников:', idGroup)
print('Лидер по количеству рецептов - это', idGroup['contributor_id'][idGroup['size'].idxmax()])
```

4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы? Обратите внимание, что отзыв с нулевым рейтингом или не заполненным текстовым описанием не считается отсутствующим.

```python
result = reviews.groupby('recipe_id').agg({'rating': ['mean']})
display(result)
print('Рецептов без отзывов:', (len(recipes)-len(result)))
```

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. 5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
revGroup = reviews.groupby(reviews['recipe_id'].tolist(), as_index= False).size()
revGroup.rename(columns = {'index':'recipe_id', 'size':'review_count'}, inplace = True)
rev = pd.merge(recipes, revGroup, on = None, left_on = 'id', right_on = 'recipe_id', how ='outer')
rev = rev.drop(['recipe_id', 'minutes','contributor_id', 'submitted', 'n_steps', 'description', 'n_ingredients', 'description_length',
              'name_word_count'], axis=1)
rev.rename(columns = {'id':'recipe_id'}, inplace = True)
rev = rev[['recipe_id', 'name', 'review_count']]
rev['review_count'].fillna(value = 0, inplace = True)
display(rev[rev['recipe_id'] == 2691])
display(rev)
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
reviews['date']= pd.to_datetime(reviews['date'])
r = reviews.groupby(pd.Grouper(key='date', freq='Y')).mean()
yr = r['rating'].idxmin()
minRateYear = yr.year
print(minRateYear)
```
